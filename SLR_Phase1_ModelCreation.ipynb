```python
import pandas as pd # pandas is a library used to access to db 
```

```python
dataset=pd.read_csv("Salary_data.csv") # read_csv function used to read the data from csv files which is a member function of pd
```

```python
dataset
```

```python
independent=dataset[["Years Experience"]]#assigned input value to independent variable from dataset
```

```python
independent
```

```python
dependent=dataset[["Salary"]]#assigned output value to dependent from dataset
```

```python
dependent
```

```python
from sklearn.model_selection import train_test_split #included train_test_split from sklearn and model selection library
X_train,X_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression() #assigned linearRegression class to an instance regressor
regressor.fit(X_train,y_train) #calling fit method of LinearRegression class to calculate weight and bias.
```

```python
weight=regressor.coef_ #assigning weight value to the variable weight
weight
```

```python
bias=regressor.intercept_ #assigning bias value to the variable bias
bias
```

```python
y_predict=regressor.predict(X_test) #predicting dependent value by passing X_test value to the function predict()
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_predict) #finding r2  value using the function r2_score()
r_score
```

```python
import pickle #to save model
filename="Finalized_model_linear.sav" #pickle extension to save model is .sav
```

```python
pickle.dump(regressor,open(filename,'wb'))
```

```python
loaded_model=pickle.load(open("Finalized_model_linear.sav",'rb'))
result=loaded_model.predict([[15]])
```
